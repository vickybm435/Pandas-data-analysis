# group by using pandas 

```python
for idx,data in df.groupby(["gear"]):
    print("ID of the group ::: ",idx)


    print("data of the group ==> \n",data)
    print("=="*20)
```

```python
for idx,data in df.groupby(["gear","cyl"]):
    print("ID of the group ::: ",idx)


    print("data of the group ==> \n",data)
    print("=="*20)
```

```python
df.groupby(["gear"])["mpg"].mean()
```

```python
df.groupby(["gear","cyl"])["mpg"].describe()
```

```python
df.groupby(["gear","cyl"])[["mpg"]].agg({"mean","count","std","var","max","median","quantile"})


```

# how to add / delete a col

```python
df["new_col"]=1
```

```python
df.head()
```

```python
#add a new column
l=[]
for i in df["gear"]:
    if i%2==0:
        l.append("even")
    else:
        l.append("odd")
df["new_col"] =l   
df
```

```python
l=[]
for i in df["vs"]:
    if i==0:
        l.append("zero")
    else:
        l.append("non_zero")
df["new_col2"] =l   
df
```

```python
df["new_col3"]=df["mpg"]>=20
```

```python
df
```

```python
df.drop(["new_col","new_col2","new_col3"],axis=1,inplace=True)
```

```python
df.drop([0],axis=0,inplace=True)
```
